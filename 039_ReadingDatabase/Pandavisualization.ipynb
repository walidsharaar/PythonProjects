```python
import matplotlib.pyplot as plt
sales = df1['sales']
months = df1['month']
plt.bar(months,sales)
plt.ylabel('Sales')
plt.xlabel('Summer Months')
plt.show()
```

```python
df2 = df.sample(1000)
```

```python
sales = df2['sales']
months= df2['month']
plt.bar(months,sales)
plt.xticks(months, rotation = 'horizontal', size = 10)
plt.ylabel('Sales')
plt.xlabel('Summer Months')
plt.show()
```
